```python
import requests, os, random
from tqdm import tqdm

os.makedirs("coco_images", exist_ok=True)

ids = random.sample(range(1, 123000), 200)

for img_id in tqdm(ids):
    url = f"http://images.cocodataset.org/train2017/{img_id:012d}.jpg"
    r = requests.get(url)
    if r.status_code == 200:
        with open(f"coco_images/{img_id:012d}.jpg", "wb") as f:
            f.write(r.content)
```
